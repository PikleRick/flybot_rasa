```python
import pandas as pd
import json
import itertools
import numpy as np
```

```python
def combinaisons(L):
    l = []
    for i in L :
        tmp = ""
        for j in i :
            tmp = tmp + " " + j
        tmp = tmp[1:]
        l.append(tmp)   
    return l
def acces_list_tuple(L):
    l = []
    for i in L :
        l.append(i[0])
    return l
```

```python
#liste des entités et des parties de la phrases
debut_phrase = ["Je veux aller", "Je souhaite partir", "Je veux un vol", "Je pars", 
           "je souhaite un vol", "j'aimerai trouver un vol","j'aimerai voyager", "cherche moi un billet", "trouve moi un vol","Je veux partir",
           "je souhaite un billet d'avion","un billet", "vol"]
connecteurs1 = ["de" , "à partir", "depuis", "en allant de"]
connecteurs2 = ["vers", "à", "en direction"]
passengers = [("pour une personne",1), ("pour 1 personne",1), ("pour deux personnes",2), ("pour 2 personnes",2), ("avec ma copine",2), ("avec ma femme",2),
              ("avec mon homme",2), ("avec 3 de mes amis",4), ("avec trois de mes amis",4), ("en couple",2), ("tout seul",1), ("solo",1), ("pour 3 personnes",3),
              ("pour 2 adultes et 2 enfants",4),("pour 2 adultes",2),("pour cinq adultes",5)]
villes = ["paris","lyon","londres","liverpool","berlin","torino","naples","rabat","cairo"]
#construire une combinaisons des mots de chaque liste
d = list(itertools.product(debut_phrase, connecteurs1, villes, connecteurs2, villes, acces_list_tuple(passengers)))
#metre lles conbinaisons sous formes de phrases en text (string)
phrases = combinaisons(d)
```

```python
cpt = 0
common_examples = {}
common_examples['common_examples'] = []
for i in debut_phrase:
    for c1 in connecteurs1:
        for vo in villes:
            for c2 in connecteurs2:
                for vd in villes:
                    for p in passengers:
                        # construire les entités de chaque phrase 
                        entites = {}
                        entites['entities'] = []
                        entites['entities'].append({
                            "start": len(i) + len(c1) + 2,
                            "end": len(vo) + len(i) + len(c1) + 2,
                            "value": vo,
                            "entity": "origin"
                        })
                        entites['entities'].append({
                            "start": len(vo) + len(i) + len(c1) + len(c2) + 4,
                            "end": len(vd) + len(vo) + len(i) + len(c1) + len(c2) + 4,
                            "value": vd,
                            "entity": "destination"
                        })
                        entites['entities'].append({
                            "start": len(vd) + len(vo) + len(i) + len(c1) + len(c2) + 6,
                            "end": len(p) + len(vd) + len(vo) + len(i) + len(c1) + len(c2),
                            "value": str(p[1]),
                            "entity": "nb_passengers"
                        })
                        # définir l'intent et le text de la phrase
                        common_examples['common_examples'].append({
                        'text': phrases[cpt],
                        'intent': "flight_search",
                        'entities': entites['entities']
                        })
                        cpt = cpt + 1
data = {"rasa_nlu_data":common_examples}
```
